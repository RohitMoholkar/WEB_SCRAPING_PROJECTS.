```python
# Web Scraping from Multiple Pages Website Flipkart. 
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import time
from IPython.display import clear_output 

phone_name = [] 
phone_MRP = [] 
discount_on_phone = [] 
phone_curr_price =[] 
phone_ratings = []  
phone_ratedby = [] 
phone_reviwedby = []
phone_RAM_ROM = []
phone_display = []
phone_camera = []
phone_battery = [] 

for i in range(1,229):  
    
    time.sleep(2) 
    
    URL = 'https://www.flipkart.com/search?q=android+phones+under+50k&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=' +str(i)  
        
    print(URL)
    clear_output(wait=True)

    HEADERS = { 'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.82 Safari/537.36',
                'Accept-Language':'en-US, en;q=0.5' } 

    response = requests.get(URL, headers=HEADERS) 

    print(response)
    clear_output(wait=True) 

    soup = BeautifulSoup(response.content, 'html.parser') 

    info = soup.find('div', class_='_1YokD2 _2GoDe3') 
    
    temp = info.find_all('div', class_='_4rR01T') 

    for val in temp:

        phone_name.append(val.text.strip())  
    
    div_tag = info.find_all('div', class_='_25b18c')  
    
    for val in div_tag:
        
        try:
            
            data = val.find('div', class_='_30jeq3 _1_WHN1').text.strip()  
        
        except Exception:
            
            data = 'N/A' 
        
        phone_curr_price.append(data) 

        try:

            data1 = val.find('div', class_='_3I9_wc _27UcVY').text.strip()  

        except Exception:

            data1 = 'N/A' 

        phone_MRP.append(data1) 

        try:

            data2 = val.find('div', class_='_3Ay6Sb').text.strip() 

        except Exception:

            data2 = 'N/A' 

        discount_on_phone.append(data2) 
    
    div_tag = info.find_all('div', class_='col col-7-12')  

    for val in div_tag:

        try:

            data1 = val.find('div', class_='_3LWZlK').text.strip() 

        except Exception:

            data1 = 'N/A' 

        phone_ratings.append(data1) 

        try:

            data = val.find('span', class_='_2_R_DZ') 

            for value in data:

                temp_list = []

                temp_str = value.text.strip() 

                temp_str = temp_str.split() 

                for ch in temp_str:

                    first_ch = ch[0] 

                    if first_ch.isnumeric():

                        ch = ch.replace(',','') 

                        temp_list.append(ch) 

            phone_ratedby.append(temp_list[0])
            phone_reviwedby.append(temp_list[1])  
    
        except Exception:
        
            phone_ratedby.append('N/A')
            phone_reviwedby.append('N/A')
    
    temp = info.find_all('ul', class_='_1xgFaf') 

    for val in temp:

        temp_str = [] 

        data = val.find_all('li', class_='rgWa7D')  

        for ch in data[:4]:

            temp_str.append(ch.text.strip()) 

        phone_RAM_ROM.append(temp_str[0]) 
        phone_display.append(temp_str[1])
        phone_camera.append(temp_str[2])
        phone_battery.append(temp_str[3]) 

print('done') 
```

```python
print(len(phone_name)) 
print(len(phone_MRP))
print(len(phone_curr_price))
print(len(discount_on_phone)) 
print(len(phone_ratings)) 
print(len(phone_ratedby))
print(len(phone_reviwedby))
print(len(phone_RAM_ROM))
print(len(phone_display)) 
print(len(phone_camera)) 
print(len(phone_battery))  
```

```python
Flipkart_DataSet = pd.DataFrame({ 'PHONE_NAME': phone_name, 
                                  'PHONE_MRP': phone_MRP, 
                                  'PHONE_CURR_PRICE': phone_curr_price, 
                                  'DISCOUNT': discount_on_phone, 
                                  'PHONE_RATINGS': phone_ratings,
                                  'RATED_BY': phone_ratedby, 
                                  'REVIWED_BY': phone_reviwedby, 
                                  'PHONE_RAM_ROM': phone_RAM_ROM, 
                                  'PHONE_DISPLAY': phone_display, 
                                  'PHONE_CAMERA': phone_camera, 
                                  'PHONE_BATTERY': phone_battery }) 
```
